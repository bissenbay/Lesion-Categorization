```python
voxel_list = sep_list.copy()
```

```python
def calculate_voxels_in_stages():
    voxel_list = sep_list.copy()
    
    result = {}
    for count, row in enumerate(voxel_list):
        # print(row, 'this is row')

        x = row[:patients_list[count][0]]
        y = row[patients_list[count][0]:patients_list[count][0]+patients_list[count][1]]
        z = row[patients_list[count][0]+patients_list[count][1]: patients_list[count][0]+patients_list[count][1]+patients_list[count][2]]
        
        stage_1 = sum([voxel_num_per_image(e) for e in x])
        stage_2 = sum([voxel_num_per_image(e) for e in y])
        stage_3 = sum([voxel_num_per_image(e) for e in z])
        
        
        
        key = 'Patient'+str(count+1)
        val = [stage_1, stage_2, stage_3]
        result[key] = val
        
    return result

        #print('Patient'+str(count+1),stage_1, stage_2, stage_3)
```

```python
result_dict = calculate_voxels_in_stages()
```

```python
import csv
with open('number_of voxels_per_stage.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in result_dict.items():
        writer.writerow([key, value])
```
